```python
import numpy as np
```

```python
class LinearRegressionGradientDescent:
    """
    Linear regression using Gradient Descent
    Simple and multiple linear regression using NumPy
    """
    def __init__(self, learning_rate=0.01, n_iterations=1000, verbose=False):
        self.lr = learning_rate
        self.n_iter = n_iterations
        self.verbose = verbose
        self.w_ = None  # Learning weights (set after fit)
        self.b_ = None  # Learning bias (set after fit)
        self.cost_history = []  # Track cost during training

    # Private helpers
    def _predict_raw(self, X):
        return X @ self.w_ + self.b_

    def _cost(self, X, y):
        m = len(y)
        errors = self._predict_raw(X) - y
        return (1 / (2 * m)) * np.sum(errors ** 2)

    def _gradient(self, X, y):
        m = len(y)
        errors = self._predict_raw(X) - y
        dw = (1 / m) * (X.T @ errors)
        db = (1 / m) * np.sum(errors)
        return dw, db
        
    # Public API
    def fit(self, X, y):
        """
        Train model on data X with targets y
        """
        X = np.array(X, dtype = float)
        y = np.array(y, dtype = float)

        # For 1D input
        if X.ndim == 1:
            X = X.reshape(-1, 1)

        m, n = X.shape

        self.w_ = np.zeros(n)
        self.b_ = 0.0
        self.cost_history = []

        for i in range(self.n_iter):
            dw, db = self._gradient(X, y)

            self.w_ -= self.lr * dw
            self.b_ -= self.lr * db

            cost = self._cost(X, y)
            self.cost_history.append(cost)

            if self.verbose and i % 100 == 0:
                print(f"Iter {i:5d} | Cost: {cost:.6f}")

        return self

    def predict(self, X):
        if self.w_ is None:
            raise RuntimeError("Didn't fit yet. Call fit() first")

        X = np.array(X, dtype = float)
        if X.ndim == 1:
            X = X.reshape(-1, 1)

        return self._predict_raw(X)

    def score(self, X, y):
        y_pred = self.predict(X)
        ss_res = np.sum((y - y_pred) ** 2)
        ss_tot = np.sum((y - np.mean(y)) ** 2)
        
        return 1.0 - ss_res / ss_tot

    def rmse(self, X, y):
        y_pred = self.predict(X)
        return np.sqrt(np.mean((y - y_pred) ** 2))

    def mae(self, X, y):
        y_pred = self.predict(X)
        return np.mean(np.abs(y - y_pred))
```
