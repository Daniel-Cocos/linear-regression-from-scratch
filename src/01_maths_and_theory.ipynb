# What I am building
## The Goal
Given training data of imputs X and targets y - learn parameters w (weights) and b (bias). I produce a prediction ŷ that is as close as possible to the true value of y

## Three Components
### Component 1 - Prediction (Forward Pass)
ŷ = Xw + b

Input X flows through the model to produce predictions

### Component 2 - Cost Function (Error Measurement)
J(w, b) = (1/2m) * Σi (ŷi - yi)^2

Measures how strong the predictions are. The 1/2 simplifies the gradient calculation

### Component 3 - Optimization (Learning)
w <- w - α * ∂J/∂w
b <- b - α * ∂J/∂b

Iteratively adjusts parameters to reduce cost.

## Gradient with respect to w
∂J/∂w = (1/m) × X^T × (ŷ - y)
       = (1/m) × X^T × (Xw + b - y)

## Gradient with respect to b
∂J/∂b = (1/m) * Σi (ŷi - yi)
       = (1/m) * sum(ŷ - y)
